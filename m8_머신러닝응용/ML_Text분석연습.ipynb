```python
from konlpy.tag import Okt

okt = Okt()
text = '파싱은 프로그램을 compile하는 과정에서 특정 프로그래밍 언어가 제시하는 문법을 잘 지켜서 작성하였는지\
compiler가 검사하는 것이다.'
#어절단위로 나눠서 추출
pharse =okt.phrases(text) 
#명사만 추출
nounlist = okt.nouns(text)
#문장을 형태소 단위로
morphslist = okt.morphs(text)
#norm은 normalize의 약자로 문장을 정규화하는 역할,stem은 각 단어에서 어간을 추출하는 기능
pos1 = okt.pos(text,norm=True,stem=True) #정규화,원형
# 각 품사를 태깅하는 역할.
#품사 태깅:주어진 텍스트를 형태소 단위로 나누고, 나눠진 각 형태소를 그에 해당하는 품사와 함께 리스트화 하는 것
# join=True로 주면
pos2 = okt.pos(text,norm=True,stem=True,join=True)

print(pharse,'\n')
print(nounlist,'\n')
print(morphslist,'\n')
print(pos1,'\n')
print(pos2,'\n')
```
