그룹 연산
- 복잡한 데이터를 어떤 기준에 따라 여러 그룹으로 나누어서 관찰할 수 있으며 이런
방식을 분할 처리하는 것을 그룹 연산이라 한다.
- 그룹 연산은 데이터를 집계, 변환, 필터링하는데 효율적이며 판다스 groupby() 메소드를
사용한다.
- 그룹객체만들기(분할)
- 그룹 연산 메소드 (적용, 결합)

적용-결합
데이터집계(agg)
- 집계 연산을 처리하는 사용자 정의함수를 그룹 객체에 적용하려면 agg()
메소드 사용
- 모든 열에 여러 함수를 매핑:group객체.agg([함수1,함수2,함수3,..])
- 각 열마다 다른 함수를 매핑:group객체.agg({'열1':함수1,'열2':함수2,...})
    
변환(transform)
- 원소의 본래 행 인덱스와 열 이름을 기준으로 연산 결과를 반환
- 데이터 변환 연산 : group객체.transform(매핑함수)
    
필터링(filter)
- 그룹 객체에 filter() 메소드를 적용할 때 조건식을 가진 함수를 전달하면
조건이 참인 그룹만을 남김
-그룹 객체 필터링: group객체.filter(조건 함수식)

객체에 함수 매핑(apply)
- 판다스 객체의 개별 원소를 특정 함수에 일대일로 매핑.사용자가 원하는
대부분의 연산을 그룹 객체에 적용
- group객체.apply(매핑함수)



```python
#Q.최대값-최소값을 반환하는 사용자 함수를 정의하고 agg()함수에 인수로
#전달하여 그룹별로 집계하여 agg_minmax 이름으로 저장 후 처음 5개행을 
#하세요 


import warnings
warnings.filterwarnings('ignore')

titanic = sns.load_dataset('titanic')
df= titanic.loc[:,['age','sex','class','fare','survived']]
grouped= df.groupby(['class'])

def MinMax (grouped):
    return grouped.max()-grouped.min()

agg_minmax = grouped.agg(MinMax)
agg_minmax
```

```python
#Q.grouped의 모든 열에 min,max 함수를 적용하여 출력하세요abs
agg_all = grouped.agg(['min','max'])
agg_all.head()
```

```python
#Q. grouped에서 age열은 mean,fare열은 min,max 함수를 적용하여 출력
# 각 열마다 다른 함수를 매핑:
#     group객체.agg({'열1':함수1,'열2':함수2,...})
 
se = group.agg({'age':'mean','fare':['max','min']})
se
```

```python
#Q.데이터 개수가 200개 이상인 그룹만을 필터링하여 데이터 프레임으로 
#반환하여  처음 5개행을 출력하세요

grouped_filter= grouped.filter(lambda x: len(x)>=200).head()
print(grouped_filter,'\n')
grouped_filter.value_counts('class')



```

```python
# Q.age 열의 평균이 30보다 작은 그룹만을 필터링하여 데이터프레임으로
#반환하여  age_filter이름으로 저장후 끝에서 5개행 출력


age_filter = grouped.filter(lambda x: x.age.mean() <30 )

age_filter

```

```python
def z_score(x):
    return (x-x.mean())/x.std()
```

```python
#Q. 평균값에서 표춘편차의 몇 배 떨어져 있는지를 평가하는 사용자 함수를 작성
z_score = grouped.apply(lambda x: x-(x.mean())/x.std())
z_score.head()
```

```python
#.Q.위에서 구한 사용자 함수를 이용해  age칼럼을 transform()함수로 변환
age_trans = grouped.age.transform(z_score)
age_trans
```
