```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

cust_df = pd.read_csv('./dataset/santander/train_santander.csv',encoding='latin-1')
cust_df.head()
```

```python
cust_df['var3'].replace(-999999,0,inplace=True)
cust_df.drop('ID',axis=1,inplace=True)
```

```python
cust_df
```

```python
X_features =cust_df.iloc[:,:-1]
y_labels =cust_df.iloc[:,-1]
print(X_features.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_features,y_labels,test_size=0.2,random_state=0)

train_cnt = y_train.count()
test_cnt = y_test.count()
print('학습데이터세트 분포\n',y_train.value_counts()/train_cnt)
print('테스트 데이터 세트 분포\n',y_test.value_counts()/test_cnt)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
import warnings
warnings.filterwarnings('ignore')
#n_estimators는 500으로 random state 는 예저수행시 동일 예측 결과 위해 설정
xgb_clf = XGBClassifier(n_estimators=500,random_state=156)
xgb_clf.fit(X_train,y_train,early_stopping_rounds=100,eval_metric='auc',eval_set=[(X_train,y_train),(X_test,y_test)])
xgb_roc_score = roc_auc_score(y_test,xgb_clf.predict_proba(X_test)[:,1],average='macro')
print('roc_acu',xgb_roc_score)
```

```python
#최적의 하이퍼 파라미터 찾기
from sklearn.model_selection import GridSearchCV
xgb_clf = XGBClassifier(n_estimators=100)
params={'max_depth':[5,7],
       'min_child_weight':[1,3],'colsample_bytree':[0.5,0.75]}
gridcv= GridSearchCV(xgb_clf,param_grid=params,cv=3)
gridcv=gridcv.fit(X_train,y_train,early_stopping_rounds=30,eval_metric='auc',eval_set=[(X_train,y_train),(X_test,y_test)])
print('최적 파라미터',gridcv.best_params_)
xgb_roc_socre = roc_auc_score(y_test,gridcv.predict_proba(X_test)[:,1],average='macro')
print('roc auc',xgb_roc_score)
```

```python
xgb_clf = XGBClassifier(n_estimators=1000,random_state=156,learning_rate=0.02,max_depth=7,min_child_weight=1,colsample_bytree=0.75,
                       reg_alpha=0.03)
xgb_clf.fit(X_train,y_train,early_stopping_rounds=200,eval_metric='auc',eval_set=[(X_train,y_train),(X_test,y_test)])
xgb_roc_auc = roc_auc_score(y_test,xgb_clf.predict_proba(X_test)[:,1],average='macro')
print('roc_acu',xgb_roc_auc)
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt
fig,ax= plt.subplots(1,1,figsize=(8,8))
plot_importance(xgb_clf,ax=ax,max_num_features=20,height=0.4)
```
